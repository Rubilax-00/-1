# Задание 1. Подготовка датасета вакансий


Вам предлагается подготовить датасет с Data-* вакансиями к анализу

```python
import numpy as np
import pandas as pd
from ast import literal_eval

pd.set_option("display.precision", 2)
pd.options.mode.copy_on_write = True
```

```python
DATA_URL = "https://raw.githubusercontent.com/akbash-e/da-ai-course-kbsu-ncfu/main/data/"
```

```python
data = pd.read_csv(DATA_URL + "data-vacancies.csv")
data.head()
```

1. Выведите первые 5 строк датасета. Примените к датасету методы `info` и `describe`.

```python
# Вывод первых строк
data.head()


```

2. Сколько в датасете строк и столбцов?

```python
# Размеры датасета
rows, cols = data.shape
print(f"Количество строк: {rows}")
print(f"Количество столбцов: {cols}")
```

3. Есть ли в датасете пропуски?

```python
# Проверка пропусков
missing_data = data.isnull().sum()
print("Пропуски в данных:\n", missing_data)

# Есть ли пропуски?
has_missing = missing_data.any()
print(f"Есть ли пропуски в данных? {'Да' if has_missing else 'Нет'}")
```

4. Есть в датасете полные дубли? Если да, удалите их.

```python
# Проверка на дубли
duplicates_count = data.duplicated().sum()
print(f"Количество полных дублей: {duplicates_count}")

# Удаление дублей
if duplicates_count > 0:
    data = data.drop_duplicates().reset_index(drop=True)
    print("Дубли удалены.")
```

5. Какие валюты (ключ `currency` в поле `salary`) используются для заработной платы? Создайте столбец `currency`.

```python
# Преобразование строки в словарь
data['salary'] = data['salary'].apply(lambda x: literal_eval(x) if isinstance(x, str) else None)

# Создание столбца currency
data['currency'] = data['salary'].apply(lambda x: x['currency'] if isinstance(x, dict) and 'currency' in x else None)

# Уникальные значения валют
currencies = data['currency'].unique()
print("Используемые валюты:", currencies)
```

6. Создайте столбец `salary_num`. Рассчитайте его, используя столбец `salary`. Если в это столбце указана вилка зарплаты, используйте среднее значение между `from` и `to`.

```python
# Рассчет средней зарплаты
def calculate_average_salary(salary):
    if isinstance(salary, dict):
        from_salary = salary.get('from', None)
        to_salary = salary.get('to', None)
        if from_salary is not None and to_salary is not None and to_salary != 'null':
            return (from_salary + to_salary) / 2
        elif from_salary is not None:
            return from_salary
    return None

# Создание столбца
data['salary_num'] = data['salary'].apply(calculate_average_salary)

# Проверка
print(data[['salary', 'salary_num']].head())
```

7. Создайте столбец `is_python_requirement`, который принимает значение `1`, если в `key_skills` присутствует `python` (учитывайте написание в любом регистре).  Какой процент вакансий требует знания python?

```python
# Создание столбца is_python_requirement
data['is_python_requirement'] = data['key_skills'].apply(
    lambda x: 1 if isinstance(x, str) and 'python' in x.lower() else 0
)

# Расчет процента вакансий, где требуется Python
python_jobs_percent = data['is_python_requirement'].mean() * 100
print(f"Процент вакансий, где требуется Python: {python_jobs_percent:.2f}%")
```

# Задание 2. Подготовка датасета магазина

Вам предлагается подготовить датасет с товарами магазина к анализу

```python
data = pd.read_csv(DATA_URL + "store-items.csv")
data.head()
```

1. Выведите первые 5 строк датасета. Примените к датасету методы `info` и `describe`

```python
data.head()
```

2. Сколько в датасете строк и столбцов?

```python
print(f"Строк: {data.shape[0]}, Столбцов: {data.shape[1]}")
```

3. Есть ли в датасете пропуски?

```python
print(data.isnull().sum())
```

4. Есть в датасете полные дубли? Если да, удалите их.

```python
duplicates = data.duplicated().sum()
print(f"Количество полных дубликатов: {duplicates}")
data = data.drop_duplicates().reset_index(drop=True)
```

5. Используя `pd.to_datetime`, приведите столбец `month_year` к дате.

```python
data['month_year'] = pd.to_datetime(data['month_year'], errors='coerce')

# Проверим результат
print(data['month_year'].head())
```

6. Создайте столбцы `year` и `month`.

```python
data['year'] = data['month_year'].dt.year
data['month'] = data['month_year'].dt.month
```

7. Создайте столбец `category_median_price`, который будет содержать медианную цену категории (у строк с одним значением `product_category_name`) будет одно значение этого столбца.

```python
# Создание столбца с медианной ценой по категориям
data['category_median_price'] = data.groupby('product_category_name')['unit_price'].transform('median')
```

8. Сделайте поле `lag_price` - цена в предыдущем месяце. Если ее нет, оставьте поле пустым.

```python
# Сортируем данные по категории и времени
data = data.sort_values(by=['product_category_name', 'month_year'])

# Создаем столбец 'lag_price', который будет содержать цену за предыдущий месяц
data['lag_price'] = data.groupby('product_category_name')['unit_price'].shift(1)

# Проверяем результат
data[['product_category_name', 'month_year', 'unit_price', 'lag_price']].head()
```

9. Создайте поле `is_price_up`, которое принимает значение `1`, если цена выросла по сравнению с предыдущим месяцем.

```python
# Создаем столбец 'is_price_up', который будет равен 1, если цена выросла по сравнению с предыдущим месяцем
data['is_price_up'] = (data['unit_price'] > data['lag_price']).astype(int)

# Проверяем результат
data[['product_category_name', 'month_year', 'unit_price', 'lag_price', 'is_price_up']].head()
```
